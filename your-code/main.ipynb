#### 3. Display Metacritic top 24 Best TV Shows of all time (TV Show name, initial release date, metascore rating, film rating system and description) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.metacritic.com/browse/tv/'
````

````python

headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

shows = soup.find_all("div", class_="c-finderProductCard")[:24]

titles = []
release_dates = []
metascores = []
ratings = []
descriptions = []

for show in shows:
    # Title
    title_tag = show.find("h3", class_="c-finderProductCard_titleHeading")
    title = title_tag.text.strip() if title_tag else "N/A"
    titles.append(title)
    
    # Release Date
    date_tag = show.find("span", class_="u-text-uppercase")
    release_date = date_tag.text.strip() if date_tag else "N/A"
    release_dates.append(release_date)
    
    # Metascore
    metascore_tag = show.find("div", class_="c-siteReviewScore")
    metascore = metascore_tag.text.strip() if metascore_tag else "N/A"
    metascores.append(metascore)
    
    rating_container = show.find("div", class_="c-finderProductCard_meta")
    if rating_container:
        rated_span = rating_container.find("span", class_="u-text-capitalize", string="Rated")
        if rated_span:

            parent_span = rated_span.find_parent("span")
            rating_text = parent_span.get_text(strip=True)

            rating = rating_text.replace("Rated", "").strip().strip('"')
        else:
            rating = "N/A"
    else:
        rating = "N/A"
    ratings.append(rating)
    
    
    # Description
    desc_tag = show.find("div", class_="c-finderProductCard_description")
    description = desc_tag.text.strip() if desc_tag else "N/A"
    descriptions.append(description)

# Создание DataFrame
df = pd.DataFrame({
    "Title": titles,
    "Release Date": release_dates,
    "Metascore": metascores,
    "Rating": ratings,
    "Description": descriptions
})


df['Title'] = df['Title'].str.replace(r'^\d+\.\s*', '', regex=True)

df
````
